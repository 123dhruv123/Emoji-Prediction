```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv("emoji-predictor-keras/train_dataset.csv",header=None)
test=pd.read_csv("emoji-predictor-keras/test_emoji.csv",header=None)
```

```python
x_train=train[0].values
y_train=train[1]

x_test=test[0].values
y_test=test[1]
```

```python
import random
combined=list(zip(x_train,y_train))
np.random.shuffle(combined)
x_train,y_train=zip(*combined)
X_train=list(x_train)
```

```python
f=open("glove_6B_50d.txt",encoding='utf-8')
```

```python
embedding_dict={}
for line in f:
    values=line.split()
    words=values[0]
    coef=np.asarray(values[1:],dtype='float32')
    embedding_dict[words]=coef
```

```python
embedding_dict['eat']
```

```python
embedding_dim=embedding_dict['eat'].shape[0]
print(embedding_dim)
```

```python
def embedding_output(X):
    max_len=10   #maxlen it will se in sentence x and y
    output=np.zeros((X.shape[0],max_len,embedding_dim))
    for ix in range(X.shape[0]):
        X[ix]=X[ix].split()   #splitting into words
        for iy in range(len(X[ix])):
            try:
                output[ix][iy]=embedding_dict[X[ix][iy].lower()]
            except:
                output[ix][iy]=np.zeros((50,))
    return output
```

```python
embedding_matrix_train=embedding_output(x_train)
embedding_matrix_test=embedding_output(x_test)
```

```python
print(x_train[0])
print(len(x_train[0]))
```

```python
print(embedding_matrix_train.shape)
print(embedding_matrix_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import *
```

```python
model=Sequential()
model.add(LSTM(64,input_shape=(10,50),return_sequences=True))
model.add(Dropout(0.5))
model.add(LSTM(64,return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(10))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
from keras.utils import to_categorical
```

```python
y_train=to_categorical(y_train,num_classes=10)
y_test=to_categorical(y_test,num_classes=10)
```

```python
from keras.callbacks import EarlyStopping,ModelCheckpoint
checkpoint=ModelCheckpoint('best_model.h5',verbose=True,monitor='val_loss',save_best_only=True)
early=EarlyStopping(monitor='val_accuracy',patience=10)
```

```python
hist=model.fit(embedding_matrix_train,y_train,shuffle=True,validation_split=0.2,epochs=100,batch_size=64)
```

```python
model.load_weights('best_model.h5')
```

```python
model.evaluate(embedding_matrix_test,y_test)
```
